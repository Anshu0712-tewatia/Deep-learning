XOR Implementation with Step Activation Aand Sigmoid

```python
import numpy as np
def step(z):
    return 1 if z >= 0 else 0
def xor_step(x1, x2):
    # Hidden layer
    h1 = step(1*x1 + 1*x2 - 0.5)
    h2 = step(-1*x1 - 1*x2 + 1.5)
    # Output layer
    y = step(1*h1 + 1*h2 - 1.5)
    return h1, h2, y
print("STEP ACTIVATION OUTPUT")
for x1 in [0,1]:
    for x2 in [0,1]:
        h1, h2, y = xor_step(x1, x2)
        print(f"x1={x1}, x2={x2} -> h1={h1}, h2={h2}, XOR={y}")
print("Code run by Anshu , 1/12/SET/BCS/445")
```

```python
import numpy as np
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
def xor_sigmoid(x1, x2):
    # Hidden layer
    h1 = sigmoid(10*x1 + 10*x2 - 5)
    h2 = sigmoid(-10*x1 -10*x2 + 15)
    # Output layer
    y = sigmoid(10*h1 + 10*h2 - 15)
    return h1, h2, y
print("\nSIGMOID ACTIVATION OUTPUT")
for x1 in [0,1]:
    for x2 in [0,1]:
        h1, h2, y = xor_sigmoid(x1, x2)
        print(f"x1={x1}, x2={x2} -> h1={h1:.3f}, h2={h2:.3f}, XOR={y:.3f}")
print("Code run by Anshu , 1/12/SET/BCS/445")
```
