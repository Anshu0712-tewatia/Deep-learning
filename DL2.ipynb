```python
import numpy as np
# for AND gate (input/output)
x_list = np.array([[0,0],[0,1],[1,0],[1,1]])
y_list = np.array([0,0,0,1])
epochs = 4
w,b = np.zeros(2),0

for epoch in range(epochs):
  for x,y in zip(x_list,y_list):
    z = np.dot(x,w) + b
    y_pred = 1 if z >= 0 else 0
    E = y - y_pred
    print("error:",E,"<- at  epoch",epoch)
    w += E*x
    b += E
    print(w,b)
print("code run by anshu 1/23/SET/BCS/445")
```

```python
# add logic to stop the trainingwhen the error is 'zero' for all the inputs.
import numpy as np
x_list = np.array([[0,0],[0,1],[1,0],[1,1]])
y_list = np.array([0,0,0,1])
epochs = 4
w,b = np.zeros(2),0

for epoch in range(epochs):
  total_error = 0
  for x,y in zip(x_list,y_list):
    z = np.dot(x,w) + b
    y_pred = 1 if z >= 0 else 0
    E = y - y_pred
    total_error += abs(E)
    print("error:",E,"<- at  epoch",epoch)
    w += E*x
    b += E
    print(w,b)

  if total_error == 0:
    print("\nNetwork converged at epoch", epoch+1)

print("code run by anshu 1/23/SET/BCS/445")
```
